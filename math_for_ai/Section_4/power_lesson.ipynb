# 累乗と平方根
累乗と平方根は、数式を記述するのに活躍します。

### 累乗とは？

同じ数または文字を何度か掛け合わせることを、累乗といいます。  
例えば、
$$3\times 3\times 3\times 3\times 3$$
は、3を5つかけていますが、次のように表すことができます。

$$3^5$$

これは、3の5乗と読みます。  
以上を踏まえて、$x$、$y$、を変数、$a$を定数として次のように表すことができます。
$$y=a^x$$
このとき、$x$、$y$、$a$は整数だけではなく小数になることもあります。

また、累乗に関して、以下の関係が成り立ちます。  
$$(x^a)^b=x^{ab}$$
$$x^ax^b=x^{a+b}$$

### 累乗をコードで実装
数式$y=a^x$をコードで実装します。

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    a = 3 
    return 3**x  # 3^x

x = np.linspace(0, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.show()
```

### 平方根とは？

以下の関係を考えます。
$$y=x^a$$
このとき、$(x^a)^b=x^{ab}$により、$a=\frac{1}{2}$であれば、$y$を2乗すると$x$になります。

$$ \begin{aligned} \\
y^2 & = (x^{\frac{1}{2}})^2 \\
& = x
\end{aligned} $$

このような、2乗して$x$になる$y$のことを、$x$の**平方根**といいます。  
平方根は次のように記述することも多いです。

$$y=\sqrt x$$

### 平方根をコードで実装
数式$y=\sqrt x$をコードで実装します。

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(x)  # xの平方根

x = np.linspace(0, 9)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.show()
```

### 演習:
コードを補完し、以下の数式のグラフを描画しましょう。
$$y=2^x$$
$$y=\sqrt x + 1$$

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(x) + 1  # 数式のコード

x = np.linspace(0, 4)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.show()
```
